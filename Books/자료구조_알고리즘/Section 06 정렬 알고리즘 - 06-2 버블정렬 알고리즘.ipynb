# 06-1 정렬 알고리즘

- 데이터를 일정한 순서로 정렬하는 정렬 알고리즘

## 정렬이란?

- 정렬(sort)이란? 키(key)를 항목값의 대소 관계에 따라 데이터 집합을 일정한 순서로 바꾸어 늘어놓는 작업

- 데이터를 정렬하면 쉽게 검색할 수 있음



### 정렬 알고리즘의 안정성

- 대표적인 알고리즘 8가지

- 정렬 알고리즘은 안정적인(stable)알고리즘과 그렇지 않은 알고리즘으로 나눌 수 있음

- 안정적 알고리즘은 값이 같은 원소의 순서가 정렬한 후에도 유지되는 것을 말함
- 안정적이지 않은 알고리즘은 정렬한 후에도 원래의 순서가 유지된다는 보장을 할수 없음

### 내부 정렬과 외부 정렬

- 카드를 최대 30장까지 한 줄로 늘어놓을 수 있는 책상에서 정렬하는 것을 생각해 보면,
  
  만약 카드가 30장 이하라면 모든 카드를 책상 위에 늘어놓고 작업을 수행 할 수 있음
  

- 카드가 500장이라면? 
  
  다 늘어놓을 수 없으므로 더 큰 책상을 마련해야 함


- 이처럼 정렬 알고리즘도 하나의 배열에서 작업할 수 있는 경우 내부 정렬(Internal sorting)을 사용하고,
   
  그렇지 않은 경우에는 외부 정렬(External sorting)을 사용
   
   
   

### 내부 정렬 : 정렬할 모든 데이터를 하나의 배열에 저장할 수 있는 경우에 사용하는 알고리즘

### 외부 정렬 : 정렬할 데이터가 많아서 하나의 배열에 저장할 수 없는 경우에 사용하는 알고리즘 

 - 외부 정렬은 내부 정렬을 응용한 것으로, 
 
   외부 정렬을 구현하려면 별도로 작업용 파일이 필요하고 알고리즘도 복잡
   

### 정렬 알고리즘의 핵심은 교환 * 선택 * 삽입

# 06-2 버블 정렬

 - 버블 정렬(buble sort)은 이웃한 두 원소의 대소 관계를 비교하여 필요에 따라 교환을 반복하는 알고리즘
 
 - 단순 교환 정렬이라고도 함

## 버블 정렬 알아보기

배열을 이용하여 버블 정렬 과정을 알아보기


```
                    v  v
초기 상태 |6|4|3|7|1|9|8|

```

- 오른쪽 끝에 있는 두 원소 9와 8에 주목

- 배열을 오름차순으로 정렬한다면 왼쪽의 값(9)이 오른쪽의 값(8)과 같거나 작어야 함

- 따라서 9와 8을 교환하면 다음과 같이 됨

```
          v v  
 |6|4|3|7|1|8|9|

```

- 다음 비교 대상인 1과 8에 주목

- 1은 8보다 작으므로 교환할 필요가 없음

- 이렇게 이웃한 원소를 비교하고 필요하면 교환하는 전체 작업

- 이때 원소 수가 n인 배열에서 n - 1 번 비교 * 교환을 하면 가장 작은 원소인 1이 맨 앞으로 이동

- 이러한 일련의 비교 * 교환하는 과정을 패스(pass)라고 함

```
           v  v - 비교 교환
|6|4|3|7|1|9|8|


         v v    - 비교 교환x
|6|4|3|7|1|8|9|

       v v      - 비교 교환
|6|4|3|7|1|8|9|


     v v        - 비교 교환
|6|4|3|1|7|8|9|


   v v          - 비교 교환
|6|4|1|3|7|8|9|

 v v            - 비교 교환
|6|1|4|3|7|8|9|

 v v            - 비교 교환
|1|6|4|3|7|8|9|

```

- 첫 번째 패스로 가장 작은 원소의 정렬이 끝났음

- 이어서 그 다음 작은 원소를 정렬하기 위해 비교 * 교환 하는 패스


```
           v v - 비교 교환x
|1|6|4|3|7|8|9|

         v v   - 비교 교환x
|1|6|4|3|7|8|9|


       v v     - 비교 교환x
|1|6|4|3|7|8|9|

     v v       - 비교 교환
|1|6|4|3|7|8|9|


   v v         - 비교 교환
|1|6|3|4|7|8|9|


|1|3|6|4|7|8|9|


```

- 이 패스를 종료하면 3은 맨 앞에서 두번째 위치로 이동그 결과 1과 3 원소의 정렬이 끝남

- 패스를 한 번 수행할 때마다 정렬할 대상은 1개씩 줄어듬

- 두 번째 패스의 비교 횟수는 첫 번째 패스보다 1번 적은 n - 2번

- 패스를 k번 수행하면 맨 앞부터 k개 원소가 정렬됨

- 모든 정렬이 끝나려면 패스를 n - 1번 수행해야 함

### 버블 정렬 프로그램

````python
# 버블 정렬 알고리즘 구현하기

from typing import MutableSequence

def bubble_sort(a: MutableSequence) -> None:
    """버블 정렬"""
    n = len(a)
#     print(f"정렬할 MutableSequence의 길이 {n}")
    
    for i in range(n - 1):
#         print(f"i = ", i)
        # 패스
        for j in range(n-1, i, -1):
#             print(f"{j-1} 인덱스와 {j} 인덱스 비교 ")
#             print(a)
            if a [j - 1] > a[j]:
#                 print(f"{j-1} 인덱스와 {j} 인덱스를 교환 ")
                a[j - 1], a[j] = a[j], a[j - 1]
#                 print(a)
                
if __name__ == "__main__":
    print("버블 정렬을 수행합니다.")
    num = int(input("원소 수를 입력하세요.: "))
    x = [None] * num # 원소 수가 num인 배열을 생성
    
    
    for i in range(num):
        x[i] = int(input(f"x[{i}]: "))
    
    print(f"정렬 전 {x}")
    bubble_sort(x) # 배열 x를 버블 정렬
    
     
    print("오름차순으로 정렬했습니다.")
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
#     print(x)
````

 - 버블 정렬은 1칸 이상 떨어져 있는 원소를 교환하는 것이 아니라 서로 이웃한 원소만 교환하므로 안정적
 
 - 원소를 비교하는 횟수는 첫 번째 패스에서는 n - 1번, 두 번째 패스에서는 n - 2번,...이므로 합계는 다음과 같음
 
(n-1) + (n-2) ..... + 1 = n(n-1) / 2

- 실제 원소를 교환하는 횟수는 배열의 원솟값에 따라 영향을 받으므로 

  그 평균값은 비교 횟수 전체의 절반인 n(n-1) / 4 번

### 교환 과정 출력

````python
# 버블 정렬 알고리즘 구현하기(정렬 과정을 출력)

from typing import MutableSequence

def buble_sort_verbose(a: MutableSequence) -> None:
    """버블 정렬(정렬 과정을 출력)"""
    
    ccnt = 0 # 비교 횟수
    scnt = 0 # 교환 횟수
    
    n = len(a)
    
    for i in range(n - 1):
        print(f'패스 {i + 1}')
        for j in range(n-1, i, -1):
            for m in range(0, n - 1):
                print(f'{a[m]:2}' + (' 'if m != j - 1 else ' +' if a[j-1] > a[j] else ' -'), end='')
            
            print(f"{a[n -1]:2}")
            ccnt += 1
            if a[j - 1] > a[j]:
                scnt += 1
                a[j - 1], a[j] = a[j], a[j -1]
                
        for m in range(0, n -1):
            print(f'{a[m]:2}', end='  ')
        print(f'{a[n-1]:2}')
        
    print(f"비교를 {ccnt}번 했습니다.")
    print(f'교환을 {scnt}번 했습니다.')
    
    
if __name__ == "__main__":
    print("버블 정렬을 수행합니다.")
    num = int(input("원소 수를 입력하세요.: "))
    x = [None] * num # 원소 수가 num인 배열을 생성
    
    
    for i in range(num):
        x[i] = int(input(f"x[{i}]: "))
    
    print(f"정렬 전 {x}")
    buble_sort_verbose(x) # 배열 x를 버블 정렬|
    
     
    print("오름차순으로 정렬했습니다.")
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
        
````

- 네 번째 패스를 보여줌 

- 여기서는 원소 교환이 한 번도 발생하지 않음

- 세 번째 패스에서 이미 모든 원소의 정렬이 완료되었기 때문

- 이미 정렬을 마친 상태라면 그 이후의 패스는 원소 교환을 하지 않음

- 다섯 번째 패스, 여섯 번째 패스에서도 원소 교환은 이루어지지 않음


- 어떤 패스의 원소 교환 횟수가 0이면 모든 원소가 정렬을 완료한 경우이므로

  그 이후의 패스는 불필요하다고 판단 정렬을 중단
  

- 처음에 주어진 배열이 다음 그림과 같이 이미 정렬을 마쳤다고 하면

  이 배열은 첫 번째 패스부터 한 번도 교환하지 않으므로 첫 번째 패스를 마친 시점에 정렬을 종료할 수 있음
  

- 이러한 중단 방식을 적용하면 정렬을 모두 마쳤거나 

  정렬이 거의 다 된 배열에서는 비교 연산을 크게 줄어 실행 시간을 단축할 수 있음


````python
# 버블 정렬 알고리즘 구현하기(알고리즘 개선 1)
# 교환이 일어나지 않으면 정렬을 멈춤

from typing import MutableSequence

def bubble_sort(a: MutableSequence) -> None:
    """버블 정렬(교환 횟수에 따른 중단)"""
    
    ccnt = 0 # 비교 횟수
    scnt = 0 # 교환 횟수
    
    n = len(a)
    
    for i in range(n - 1):
        print(f'패스 {i + 1}')
        exchange = 0
        for j in range(n-1, i, -1):
            for m in range(0, n - 1):
                print(f'{a[m]:2}' + (' 'if m != j - 1 else ' +' if a[j-1] > a[j] else ' -'), end='')
            
            print(f"{a[n -1]:2}")
            ccnt += 1
            if a[j - 1] > a[j]:
                scnt += 1
                a[j - 1], a[j] = a[j], a[j -1]
                exchange += 1
        print(f"{i+1} 패스 교환 횟수 {exchange}")
                
        if exchange == 0:
            print("교환이 없어서 종료")
            break

        for m in range(0, n -1):
            print(f'{a[m]:2}', end='  ')
        print(f'{a[n-1]:2}')
        
    print(f"비교를 {ccnt}번 했습니다.")
    print(f'교환을 {scnt}번 했습니다.')
    
    
if __name__ == "__main__":
    print("버블 정렬을 수행합니다.")
    num = int(input("원소 수를 입력하세요.: "))
    x = [None] * num # 원소 수가 num인 배열을 생성
    
    
    for i in range(num):
        x[i] = int(input(f"x[{i}]: "))
    
    print(f"정렬 전 {x}")
    bubble_sort(x) # 배열 x를 버블 정렬
    
     
    print("오름차순으로 정렬했습니다.")
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
        
````

### 알고리즘 개선 2

   v v          - 비교 교환 x

- 마지막 교환한 위치에서 3개의 원소 1,3,4는 정렬된 상태

- 이렇게 각각의 패스에서 비교 교환을 하다가 어떤 특정한 원소 이후에 교환하지 않는다면

  그 원소보다 앞쪽에 있는 원소는 이미 정렬을 마친 것
  
- 따라서 두 번째 패스는 첫 번째 원소를 제외한 6개가 아니라 4개로 좁혀서 비교 교환


```

 - - - 

         v v    - 비교 교환x

 - - - 


       v v      - 비교 교환
|1|3|4|9|6|7|8|
 - - - 

      
|1|3|4|6|9|7|8|
 - - - - 



```

````python
from typing import MutableSequence

def bubble_sort(a: MutableSequence) -> None:
    """버블 정렬(스캔 범위를 제한)"""
    
    n = len(a)
    
    k = 0
    
    while k < n - 1 :
        last = n - 1
        for j in range(n -1, k, -1):
            if a[j - 1] > a[j]:
                a[j-1], a[j] = a[j], a[j -1]
                # 마지막 교환 위치
                last = j
                
        k = last
````

````python
# 버블 정렬 알고리즘 구현하기(알고리즘 개선 2)
# 마지막 교환 위치를 저장하여 더 이상 교환이 일어나지 않는다면
# 마지막 교환 위치에서부터 스캔함
from typing import MutableSequence

def bubble_sort3_verbose(a: MutableSequence) -> None:
    """버블 정렬(스캔 범위를 제한)"""
    ccnt = 0  # 비교 횟수
    scnt = 0  # 교환 횟수
    n = len(a)
    k = 0
    i = 0
    while k < n - 1:
        print(f'패스 {i + 1}')
        i += 1
        last = n - 1
        for j in range(n - 1, k, -1):
            for m in range(0, n - 1):
               print(f'{a[m]:2}' + ('  ' if m != j - 1 else
                                    ' +' if a[j - 1] > a[j] else ' -'),
                     end='')
            print(f'{a[n - 1]:2}')
            ccnt += 1
            if a[j - 1] > a[j]:
                scnt += 1
                a[j - 1], a[j] = a[j], a[j - 1]
                last = j
        k = last
        for m in range(0, n - 1):
           print(f'{a[m]:2}', end='  ')
        print(f'{a[n - 1]:2}')
    print(f'비교를 {ccnt}번 했습니다.')
    print(f'교환을 {scnt}번 했습니다.')

if __name__ == '__main__':
    print('버블 정렬을 수행합니다')
    num = int(input('원소 수를 입력하세요.: '))
    x = [None] * num  # 원소 수 num인 배열을 생성

    for i in range(num):
        x[i] = int(input(f'x[{i}] : '))

    bubble_sort3_verbose(x)  # 배열 x를 버블

    print('오름차순으로 정렬했습니다.')
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
````

### 셰이커 정렬 알아보기

- 기본 버블 정렬,개선1, 개선2를 실행하면 모두 비교를 21번 하고 교환은 6번함

- 홀수 패스에서는 가장 작은 원소를 맨 앞으로 이동시키고,
  
  짝수 패스에서는 가장 큰 원소를 맨 뒤로 이동시켜 패스의 이동방향을 번갈아 바꾸어 봄
  
- 그러면 앞의 실행 결과보다 적은 비교 횟수로 정렬할 수 있음

- 이렇게 버블 정렬을 개선한 알고리즘을 셰이커(Shaker sort)이라고 하며,
  
  양방향 버블 정렬(bidirectional bubble sort), 칵테일 정렬(cocktail sort), 칵테일 셰이커 정렬(cocktail shaker sort)라고도 함

````python
# 셰이커 정렬 알고리즘 구현하기

from typing import MutableSequence

def shaker_sort(a: MutableSequence) -> None:
    """셰이커 정렬"""
    
    left = 0
    right = len(a) - 1
    last = right
    
    while left < right:
        for j in range(right, left, -1):
            if a[j - 1] > a[j]:
                a[j - 1], a[j] = a[j] , a[j - 1]
                last = j
        left = last
        
        for j in range(left, right):
            if a[j] > a[j + 1]:
                a[j], a[j+1] = a[j+1], a[j]
                last = j
        right = last
````

````python
from typing import MutableSequence

def shaker_sort_verbose(a: MutableSequence) -> None:
    """"셰이커 정렬(정렬 과정을 출력)"""
    ccnt = 0  # 비교 횟수
    scnt = 0  # 교환 횟수
    left = 0
    n = len(a)
    right = len(a) - 1
    last = right
    i = 0
    while left < right:
        print(f'패스{i + 1}')
        i += 1
        for j in range(right, left, -1):
            for m in range(0, n - 1):
               print(f'{a[m]:2}' + ('  ' if m != j - 1 else
                                    ' +' if a[j - 1] > a[j] else ' -'),
                     end='')
            print(f'{a[n - 1]:2}')
            ccnt += 1
            if a[j - 1] > a[j]:
                scnt += 1
                a[j - 1], a[j] = a[j], a[j - 1]
                last = j
        left = last
        for m in range(0, n - 1):
           print(f'{a[m]:2}', end='  ')
        print(f'{a[n - 1]:2}')

        if (left == right):
             break
        print(f'패스 {i + 1}')
        i += 1
        for j in range(left, right):
            for m in range(0, n - 1):
               print(f'{a[m]:2}' + ('  ' if m != j else
                                    ' +' if a[j] > a[j + 1] else ' -'),
                     end='')
            print(f'{a[n - 1]:2}')
            ccnt += 1
            if a[j] > a[j + 1]:
                scnt += 1
                a[j], a[j + 1] = a[j + 1], a[j]
                last = j
        right = last
        for m in range(0, n - 1):
           print(f'{a[m]:2}', end='  ')
        print(f'{a[n - 1]:2}')
    print(f'비교를 {ccnt}번 했습니다.')
    print(f'교환을 {scnt}번 했습니다.')

if __name__ == '__main__':
    print('셰이커 정렬을 수행합니다.')
    num = int(input('원소 수를 입력하세요.: '))
    x = [None] * num  # 원소 수 num인 배열을 생성

    for i in range(num):
        x[i] = int(input(f'x[{i}] : '))

    shaker_sort_verbose(x)  # 배열 x를 단순 교환 정렬

    print('오름차순으로 정렬했습니다.')
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
````
