# 10.1 퀵 정렬 알고리즘(Quck Sort Algorithm)

 - 퀵 정렬 알고리즘은 이름에서도 알 수 있듯이 현존하는 정렬 알고리즘 중에서 가장 빠르다고 알려져 있는 정렬 알고리즘
 
 - 다른 알고리즘보다 속도 면에서 월등하게 우수

 - 퀵 정렬 알고리즘은 한 번의 반복만을 사용

```
# 정렬 전 데이터
|1|4|2|5|7|3|6|8|

```

- 정렬 되어 있지 않은 데이터들이 위와 같이 8개 있음

- 퀵 정렬 알고리즘의 기본 개념은 이중에 하나를 선택해서 이것을 기준으로 양쪽으로 편을 가름

- 문제는 그 하나를 어떻게 선택하느냐가 문제이고 또 그 선택 방법에 따라서 퀵 정렬 알고리즘의 전체 성능을 좌우함

- 편의상 가운데 있는 하나를 선택해서 기준을 선택


```
# 정렬 전 데이터

       v 기준으로 선택 5
|1|4|2|5|7|3|6|8|

```

- 정 가운데를 기준이 되는 데이터로 선택하고 기준 데이터의 왼쪽부분을 왼쪽 팀, 오른쪽 부분을 오른쪽 팀으로 설정

```
# 정렬 전 데이터

       v 
|1|4|2|5|7|3|6|8|
 -             -


```
 
 - 왼쪽 팀의 가장 선두를 왼쪽 팀장
 - 오른쪽팀의 가장 오른쪽 데이터를 오른쪽 팀장
 
 - 왼쪽과 오른쪽 팀의 데이터와 기준이 되는 가운데 데이터를 비교해서 기준이 되는 데이터보다작으면 왼쪽으로 이동
 
 - 기준이 데이터보다 크면 오른쪽 팀으로 이동

```
       v 
|1|4|2|5|7|3|6|8|
 -             -

       v 
|1|4|2|5|7|3|6|8|
   -         -

       v           - 3 왼쪽으로 이동
|1|4|2|5|7|3|6|8|
     -     -

         v 
|1|4|2|3|5|7|6|8|
     -     -

# 한번의 정렬이 끝난 후의 모습
         v 
|1|4|2|3|5|7|6|8|
     -     -

```

- 기준이 되는 데이터의 왼쪽에는 기준 데이터보다 작은 데이터들이 모여 있게 되고,

  오른쪽에는 기준 데이터보다 큰 데이터들이 모여 있게 됨


```
     v   v   v 
|1|4|2|3|5|7|6|8|

- 새로 선택된 중간값 2, 6
```

- 왼쪽 팀에서 다시 가운데를 기준이 되는 데이터로 선택하고 오른쪽도 마찬가지로 가운데를 기준이 되는 것으로 선택


- 그 다음부터는 동일

- 왼쪽 팀에서 기준이 되는 데이터는 '2'가 되며 

  그 왼쪽과 오른쪽 데이터들을 기준이 되는 데이터 '2'와 비교하여 '2'보다 작은 데이터를 왼쪽, 큰쪽 데이터를 오른쪽으로 이동
  
- 오른쪽 팀도 마찬가지로 '6'과 비교해서 더 큰 데이터를 '6'의 오른쪽 작은 데이터는 왼쪽으로 이동 시킴

- 이러한 과정을 전체 데이터가 모두 정렬될 때까지 반복하면 퀵 정렬 알고리즘이 종료하게 됨


````python
import random
import time

compare_counter = 0
swap_counter = 0

def swap(x, i, j):
    x[i], x[j] = x[j], x[i]

    
def pivotFirst(x, lmark, rmark):
    pivot_val = x[lmark]
    pivot_idx = lmark
    
    while lmark <= rmark:
        
        while lmark <= rmark and x[lmark] <= pivot_val:
            lmark += 1
        while lmark <= rmark and x[rmark] >= pivot_val:
            rmark -= 1
            
        if lmark <= rmark:
            swap(x, lmark, rmark)
            lmark += 1
            rmark -= 1
            
    swap(x, pivot_idx, rmark)
    return rmark


def quickSort(x, pivotMethod=pivotFirst):
    def _qsort(x, first, last):
        if first < last:
            
            splitpoint = pivotMethod(x, first, last)
            
            _qsort(x, first, splitpoint -1)
            
            _qsort(x, splitpoint+1, last)
            
    _qsort(x, 0, len(x)-1)
            
            
if __name__ == "__main__":
    list = []
    input_n = input("정렬할 데이터의 수:")
    for i in range(int(input_n)):
        list.append(random.randint(1, int(input_n)))
    print("정렬 전")
    print(list)
    
    start_time = time.time()
    quickSort(list)
    running_time = time.time() - start_time
    
    print("정렬후")
    print(list)
    print(running_time)
````

 - 퀵 정렬 함수는 기존의 다른 정렬 함수와는 달리 재귀호출을 사용함
 
 - 재귀 호출을 사용하지 않고도 퀵 정렬 함수의 구현이 가능하긴 하지만 
   
   퀵 정렬의 특성상 재귀 호출을 사용하는 것이 훨씬 간단
   
   
- 퀵 정렬 함수는 3개의 매개 변수를 사용

- 첫 번째는 정렬할 데이터들이 저장되어 있는 리스트
- 두 번째는 왼쪽 인덱스를 가르키는 first
- 세 번째는 오른쪽 인덱스를 가르키는 last



 - 퀵 정렬 함수가 실행되면 변수 first가 변수 last보다 작은가를 검사
 
   이 부분이 퀵 정렬 함수가 재귀 호출될 때 재귀 호출을 끝내는 조건이 됨
 
 - first가 last 보다 작으면 주어진 데이터를 2개의 팀으로 나누는 역할을 하는 pivotMethod() 함수를 호출하여
 
   중간값인 splitpoint를 구함
   
 - 왼쪽 팀은 first, splitpoint-1인 구간으로, 오른쪽 팀은 splitpoint+1과 last의 구간으로 새로운 정렬 범위를 설정

# 10.2 퀵 정렬 알고리즘의 분석

- 퀵 정렬 알고리즘의 성능은 O(N * log N)

- N개의 데이터가 있다고 가정
- 데이터들을 분할할 때 일반적으로 N/2 정도로 분할된다고 가정

- 소요 시간을 T(N)이라고 하면 
  재귀 호출을 하는 퀵 정렬 함수에서 첫 번째 퀵 정렬 함수가 실행되면 다음과 같은 식이 될 수 있음
  
  T(N) -> cN + 2T(N/2)  (c는 상수)
  
  
- 위 식이 의미하는 것은 N개의 데이터를 정렬하는데 걸리는 시간 T(N)은 cN이라는 상수값
  즉 한 번 퀵 정렬 함수를 실행했을 때 소요되는 정렬 시간과 그 데이터들을 2개로 쪼개어있을 때 걸ㄹ리는 시간을 더 한 값과 같거나 작다는 의미
  
- 또한 위의 식을 퀵 정렬 함수가 두 번째 실행되면 다음과 같은 식이 됨

  O(T(N)) = O(N * log N)
  
- 이것은 퀵 정렬 알고리즘의 경우 일반적인 데이터의 집합이라면 데이터를 쪼갤 때 대부분 거의 반으로 쪼개 진다는 것을 의미

- 그에 비해서 이미 데이터가 정렬되어 있는 최선의 경우나 
  
  데이터가 역으로 정렬되어 있는 최악의 경우는 오히려 성능이 많이 떨어지는 것을 알 수 있음
  
- 실제로 퀵 정렬 알고리즘이 일반적인 경우에는 O(NlogN)의 성능을 보여주지만,

  최악의 경우에는  O(N^2)의 성능이 되기 때문에 앞에서 배운 4가지 알고리즘과 별반 차이가 없게 됨
  
- 보통 제일 빠른 알고리즘이라 생각하기 쉬운데 반드시 모든 경우에 해당되는 것은 아님

### 시간의 효율성

- 퀵 정렬 알고리즘은 O 표기법에 의하면 O(NlogN)의 실행 시간을 갖음

- 다른 정렬 알고리즘들이 O(N^2)의 실행 시간을 갖는데 비해서 보면 상당히 빠른 정렬 알고리즘

- 그러나 일반적인 경우에는 빠를지 몰라도 최악의 경우와 같이 이미 정렬이 되어 있는 상태에서는 
  
  아무리 퀵 정렬 알고리즘이라고 할지라도 기존의 정렬 알고리즘과 동일하게 O(N^2)의 시간 복잡도를 갖게 됨
  
  
### 공간의 효율성

- 퀵 정렬 알고리즘의 경우는 사실 정렬을 하기 위한 별도의 저장 공간이 필요가 없음

### 코드의 효율성

- 퀵 정렬 알고리즘은 보통의 경우 재귀 호출을 사용

- 재귀 호출을 사용하게 되면 코드의 길이가 짧아지고 컴팩트하다는 장점이 있는 반면
  디버깅이 어렵고 코드를 이해하기가 좀 어려워진다는 단점이 있음

- 재귀 호출을 사용하지 않고 스택을 사용하여 퀵 정렬 알고리즘을 구현할 수도 있음

# 10.3 기수 정렬 알고리즘(Radix Sort Algorithm)

기수 정렬이란?

- 정렬할 데이터의 자릿수를 이용하여 데이터를 정렬하는 방법

- 기수 정렬 알고리즘의 기본 개념은 자릿수로 데이터를 정렬한다는 것

```

# 정렬 되지 않은 데이터

|22|35|78|45|19|72|66|82|53|55|

```

```
# 1의 자릿수로 정렬한 데이터

                  55
      82          35
  21  72  53      45  66      78  19  
  1   2   3   4   5   6   7   8   9     

```

- 이와 같이 모든 데이터들을 가장 아래의 자릿수에 따라서 정렬


- 일의 자릿수로 정렬한 데이터를 다음과 같이 일렬로 배열

```
|21|    |72|82|    |53|         |45|35|55|          |66|   |78|        |19|
```

- 이제 다시 위의 데이터들을 10의 자리로 정렬

```

                    55      78 
    19  21  35  45  53  66  72  82
    1   2   3   4   5   6   7   8   9     

```

- 다시 데이터를 하나씩 꺼내어 일렬로 만듬

```
  19     21  35     45    53  55 66     72  78  82
```

````python
from math import log10

from random import randint


def get_digit(number, base, pos):
    return (number // base**pos) % base

def prefix_sum(array):
    for i in range(1, len(array)):
        array[i] = array[i] + array[i-1]
    return array

def radixsort(l, base=10):
    passes = int(log10(max(l)) + 1)
    output = [0] * len(l)

    for pos in range(passes):
        count = [0] * base
        
        for i in l:
            digit = get_digit(i, base, pos)
            count[digit] += 1
            
        count = prefix_sum(count)
        
        for i in reversed(l):
            digit = get_digit(i, base, pos)
            count[digit] -= 1
            new_pos = count[digit]
            output[new_pos] = i
        
        l = output[:]
    
    return output

if __name__ == "__main__":
    l = []
    l = [randint(1, 99999) for x in range(100)]
    print("정렬 전")
    print(l)
    
    sorted = radixsort(l)
    print("정렬 후")
    print(sorted)
````

# 10.4 기수 정렬 알고리즘의 실행과 성능

- 기수 정렬 알고리즘의 성능은 자릿수와 정렬할 데이터의 수 그리고 각 자리수에 따른 큐의 수에 따라 좌우

- 데이터 자릿수 D와 정렬할 데이터 수 N 그리고 그에 해당하는 큐의 수 Q가 있다고 가정할 때
  다음과 같은 알고리즘 성능을 갖게 됨
  
O(RadixSort) == O(D(N + Q))

- 따라서 비교 횟수나 이동 횟수가 데이터의 정렬된 상태와 상관없이 데이터의 수 + 큐의 수와 더불어
  데이터의 자리수에 의해 영향을 받게 됨
  
- 100보다 작은 숫자의 정렬의 경우에는 자릿수 D는 1의자리, 10의 자리가 되어 총 2가 되지만
  이진수로 표현하는 경우에는 그 수가 급격히 증가되므로 알고리즘의 전체 선능이 저하되는 단점이 생김
  
- 이진수 표현법을 사용하는 것은 몯든 데이터에 대한 처리를 원활하게 하기 위한 점

- 이진수로 표현하는 경우 프로그램 코드에서 비트 연산을 사용할수 있으므로 한 번의 연산속도는 비트 연산이 훨씬 빠름

### 시간의 효율성

- 기수 정렬 알고리즘은 O(D(N + Q))의 성능

- 데이터의 수가 적거나 정렬할 데이터의 자릿수가 적은 경우 성능 향상

### 공간의 효율성

- 기수 정렬 알고리즘은 공간의 효율성이 가장 안좋은 정렬 알고리즘

- 별도의 큐를 위한 공간을 확보해야 하기 때문

- 메모리를 많이 소비하는 알고리즘은 정렬 속도가 빠르다고 해도 그다지 좋은 알고리즘이 아님

### 코드의 효율성

- 이진수로 데이터를 정렬하는 방법을 사용하면 비트 연산이 포함되기 때문에 

  코드의 내용이 상당히 어려워짐
